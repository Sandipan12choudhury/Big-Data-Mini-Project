```python
# Counting the frequency of all words present in a file ( Input file (text) uploaded from local file system )
from pyspark import SparkContext, SparkConf

conf = SparkConf().setAppName("WordCount").setMaster("local")
sc = SparkContext.getOrCreate(conf=conf)
file_path = input("Enter the path to your text file: ")
text_file = sc.textFile(file_path)
word_counts = (
    text_file
    .flatMap(lambda line: line.split())  # Split each line into words
    .map(lambda word: (word, 1))         # Map each word to a (word, 1) pair
    .reduceByKey(lambda a, b: a + b)     # Sum up counts for each word
)
for word, count in word_counts.collect():
    print(f"{word}: {count}")
sc.stop()
```

```python
# Analyzing a video stream ( detecting moving objects in the video ) : Input video uploaded from local file system
!pip install pyspark

from google.colab import files
import cv2
import os
from time import sleep
import shutil

uploaded = files.upload()
video_path = next(iter(uploaded.keys()))

frame_directory = "/content/frames/"
os.makedirs(frame_directory, exist_ok=True)

cap = cv2.VideoCapture(video_path)
count = 0

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
   frame_path = os.path.join(frame_directory, f"frame_{count}.jpg")
    cv2.imwrite(frame_path, frame)
    count += 1

cap.release()
print(f"Extracted {count} frames into '{frame_directory}'")

output_folder = "/content/output_frames/"
os.makedirs(output_folder, exist_ok=True)

def process_frame(path):
    frame = cv2.imread(path)
    if frame is None:
        print(f"Warning: Failed to read frame at {path}")
        return

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray, (5, 5), 0)
    _, thresh = cv2.threshold(blur, 20, 255, cv2.THRESH_BINARY)

    contours, _ = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

    if len(contours) == 0:
        print(f"No contours found in {path}")

    for contour in contours:
        if cv2.contourArea(contour) < 700:
            continue
        x, y, w, h = cv2.boundingRect(contour)
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

    output_path = os.path.join(output_folder, os.path.basename(path))
    cv2.imwrite(output_path, frame)
    print(f"Processed and saved: {output_path}")

frame_files = os.listdir(frame_directory)
if len(frame_files) == 0:
    print("No frames found to process. Check the 'frames' directory.")

for frame_file in frame_files:
    frame_path = os.path.join(frame_directory, frame_file)
    process_frame(frame_path)

output_files = os.listdir(output_folder)
if len(output_files) == 0:
    print(f"No processed frames found in '{output_folder}'.")
else:
    print(f"Processing complete. {len(output_files)} frames processed.")
    shutil.make_archive('/content/output_frames', 'zip', '/content/output_frames')
    print("Processed frames zipped.")

from google.colab import files
files.download('/content/output_frames.zip')
```
